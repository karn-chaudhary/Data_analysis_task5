```python
 import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic = pd.read_csv(r"C:\Users\HP\Favorites\Downloads\titanic.csv")
```

```python
titanic.head(4)
```

```python
print("Dataset Info:")
titanic.info()
```

```python
print("\nDataset Description:")
titanic.describe()
```

```python
print("\nValue Counts for each column:")
for col in titanic.select_dtypes(include=['object']).columns:
    print(f"\nValue Counts for {col}:\n", titanic[col].value_counts())
```

```python
print("\nGenerating Pairplot...")
sns.pairplot(titanic, diag_kind='kde')
plt.suptitle('Pairplot of Dataset', y=1.02)
plt.show()
```

```python

print("\nGenerating Heatmap...")
numeric_data = titanic.select_dtypes(include=['number'])

plt.figure(figsize=(12,8))
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
print("\nGenerating Histograms...")
titanic.hist(figsize=(15,10), bins=30, edgecolor='black')
plt.suptitle('Histograms of Numeric Columns')
plt.show()
```

```python
print("\nGenerating Boxplots...")
numeric_cols = titanic.select_dtypes(include=['float64', 'int64']).columns
for col in numeric_cols:
    plt.figure(figsize=(6,4))
    sns.boxplot(x=titanic[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

```python
print("\nGenerating Scatterplots...")
for i, col1 in enumerate(numeric_cols):
    for j, col2 in enumerate(numeric_cols):
        if i < j:
            plt.figure(figsize=(6,4))
            sns.scatterplot(x=titanic[col1], y=titanic[col2])
            plt.title(f'Scatterplot between {col1} and {col2}')
            plt.show()
```
